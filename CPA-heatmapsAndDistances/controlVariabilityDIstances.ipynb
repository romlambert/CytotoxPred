```python
import pandas as pd
import seaborn as sns
import scipy as sp
import numpy as np
from sklearn.decomposition import PCA
import plotly.express as px
import matplotlib.pyplot as plt

sns.set(style='white', context='notebook', rc={'figure.figsize':(12,10)}, font='sans-serif')

#Input and table cleaning
blasts = pd.read_csv("C:\\Users\\lambr\\OneDrive - National University of Ireland, Galway\\CellPainting project data\\Harmony\\Nv6-PerWell\\blastsProfiles.csv")
blasts = blasts.dropna(axis=1)
blasts.set_index(['Compound','Concentration','PlateName'],inplace=True)
blastTreat = blasts
blastTreat.drop(columns=['Unnamed: 0', 'Count'],inplace = True)
# filter only the DMSO treated wells
blastTreat = blastTreat.xs('DMSO', level='Compound')
blastTreat.reset_index(inplace=True)
blastTreat.drop(columns=['Concentration'], inplace = True)
blastTreat['wellID'] = blastTreat['PlateName'] + '_' + blastTreat['Row'].astype(str) + '_' + blastTreat['Column'].astype(str)
blastTreat.drop(columns=['PlateName','Row','Column'], inplace = True)
blastTreat.set_index('wellID', inplace=True)
blastTreat
```

```python
# compute mean profile
blastTreatMean = blastTreat.mean(axis=0)
blastTreatMean
# insert the mean profile as the first row
blastTreat = pd.concat([blastTreatMean.to_frame().T, blastTreat])
blastTreat
```

```python
### calculate inverted covariance matrix between PCA components
profMatrix = blastTreat.to_numpy()


covMat = np.cov(profMatrix, rowvar=False)
covMatInv = np.linalg.inv(covMat)
sampleSize=len(blastTreat.index)

### compute matrix of Mahalanobis distances
sample = []
mahaMat = np.empty(shape=[sampleSize,sampleSize])
for x in range(0,sampleSize):
    for y in range(0,sampleSize):
        sample.insert(y, sp.spatial.distance.euclidean(profMatrix[x],profMatrix[y])) # mahalanobis distance
    mahaMat[x] = sample
    sample = []
mahaDF = pd.DataFrame(mahaMat, index=blastTreat.index, columns=blastTreat.index)
```
