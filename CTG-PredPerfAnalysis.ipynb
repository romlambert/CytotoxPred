```python
# analysis of RF performace for viability prediction in tubes / blasts, CellProfiler/Harmony
# 30/10/2023, Roman Lambert
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# feature importances from RF models

#onedrive_path = "C:/Users/lambr/OneDrive - National University of Ireland, Galway/"
onedrive_path = "E:/OneDrive - National University of Ireland, Galway/"

viab_1 = onedrive_path + "ViabilityPred-models\\Harmony\\Blasts\\stats_df.csv"
viab_2 = onedrive_path + "ViabilityPred-models\\Harmony\\Tubes\\stats_df.csv"
viab_3 = onedrive_path + "ViabilityPred-models\\CellProfiler\\Blasts\\stats_df.csv"
viab_4 = onedrive_path + "ViabilityPred-models\\CellProfiler\\TubesMono\\stats_df.csv"

count_1 = onedrive_path + "CountPred-models\\Harmony\\Blasts\\stats_df.csv"
count_2 = onedrive_path + "CountPred-models\\Harmony\\Tubes\\stats_df.csv"
count_3 = onedrive_path + "CountPred-models\\CellProfiler\\Blasts\\stats_df.csv"
count_4 = onedrive_path + "CountPred-models\\CellProfiler\\TubesMono\\stats_df.csv"

stats_df_1 = pd.read_csv(count_1, index_col=0).sort_values(by='R2', ascending=False)
stats_df_1['Cell type'] = 'Blasts'
stats_df_2 = pd.read_csv(count_2, index_col=0)
stats_df_2['Cell type'] = 'Tubes'
stats_df = pd.concat([stats_df_1, stats_df_2], axis=0)
stats_df
```

```python
stats_df_1_viab = pd.read_csv(viab_1, index_col=0).sort_values(by='R2', ascending=False)
stats_df_1_viab['Cell type'] = 'Blasts'
stats_df_2_viab = pd.read_csv(viab_2, index_col=0)
stats_df_2_viab['Cell type'] = 'Tubes'
stats_df_viab = pd.concat([stats_df_1_viab, stats_df_2_viab], axis=0)

#plot a barplot of R2 sc
sns.set_theme(style="darkgrid")
ax = sns.barplot(x="R2", y=stats_df_viab.index, data=stats_df_viab,orient='h', hue='Cell type',  )
ax.bar_label(ax.containers[0], fontsize=8, fmt = ' %.2f')
ax.bar_label(ax.containers[1], fontsize=8, fmt = ' %.2f')
ax.figure.set_size_inches(8, 10)
plt.show()
```

```python
#plot a barplot of R2 sc
sns.set_theme(style="darkgrid")
ax = sns.barplot(x="R2", y=stats_df.index, data=stats_df,orient='h', hue='Cell type' )
ax.bar_label(ax.containers[0], fontsize=8, fmt = ' %.2f')
ax.bar_label(ax.containers[1], fontsize=8, fmt = ' %.2f')
ax.figure.set_size_inches(8, 10)
plt.show()
```
